```python
import matplotlib.pyplot as plt
from opensignalsreader import OpenSignalsReader
```

### Cargar los datos desde el archivo TXT de Bitalino

```python
data1 = OpenSignalsReader('data_reposo/BRAZO_FUERZA_I.txt')
data2 = OpenSignalsReader('data_reposo/BRAZO_FUERZA_D.txt')
data3 = OpenSignalsReader('data_estimulo/BRAZO_CANSADO_FUERZA_D.txt')
data4 = OpenSignalsReader('data_reposo/BRAZO_MOVIMIENTOSVARIOS_D.txt')
data5 = OpenSignalsReader('data_estimulo/BRAZO_CANSADO_MOVIMIENTOS_D.txt')
data6 = OpenSignalsReader('data_reposo/BRAZO_REPOSO_D.txt')
data7 = OpenSignalsReader('data_estimulo/BRAZO_CANSADO_REPOSO_D.txt')
data8 = OpenSignalsReader('data_reposo/CORAZON_REPOSO_SANO.txt')
data9 = OpenSignalsReader('data_estimulo/CORAZON_EJERCICIO-REPOSO.txt')
```

# Mediciones de fuerza

```python
plt.figure(figsize=(14, 6))

plt.plot(data1.raw([1]))
plt.xlabel('Sample')
plt.ylabel('Value')
plt.title('Left Arm Rest Contraction Data')
plt.show()
```

```python
plt.subplot(2, 1, 1)
plt.plot(data2.raw([1]))
plt.xlabel('Sample')
plt.ylabel('Value')
plt.title('Right Arm Rest Contraction Data')


plt.subplot(2, 1, 2)
plt.plot(data3.raw([1]))
plt.xlabel('Sample')
plt.ylabel('Value')
plt.title('Right Arm Tired Contraction Data')

plt.tight_layout()

plt.show()
```

```python
plt.subplot(2, 1, 1)
plt.plot(data4.raw([1]))
plt.xlabel('Sample')
plt.ylabel('Value')
plt.title('Right Arm Moving Resting Data')


plt.subplot(2, 1, 2)
plt.plot(data5.raw([1]))
plt.xlabel('Sample')
plt.ylabel('Value')
plt.title('Right Arm Moving Tired Data')

plt.tight_layout()
```

```python
plt.subplot(2, 1, 1)
plt.plot(data6.raw([1]))
plt.xlabel('Sample')
plt.ylabel('Value')
plt.title('Right Arm Resting Data')


plt.subplot(2, 1, 2)
plt.plot(data7.raw([1]))
plt.xlabel('Sample')
plt.ylabel('Value')
plt.title('Right Arm Tired Data')

plt.tight_layout()
```
