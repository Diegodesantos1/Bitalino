### Mediciones de EMG (Electromiograf√≠a)

```python
plt.figure(figsize=(14, 6))
plt.plot(data7.raw([1]))
plt.title('EMG signal')
plt.xlabel('Sample')
plt.ylabel('Value')
plt.show()
```

En esta prueba de diagn√≥stico analizamos la salud de los m√∫sculos y las c√©lulas nerviosas (neuronas) en los diferentes movimientos realizados en el brazo.

Tambi√©n en este caso vamos a ampliar la gr√°fica para poder ver mejor los datos.

```python
plt.figure(figsize=(14, 6))
plt.plot(data7.raw([1]))
plt.xlim(0, 700)
plt.show()

plt.figure(figsize=(14, 6))
plt.plot(data7.raw([1]))
plt.xlim(700, 1500)
plt.show()
```

Como podemos observar existe una diferencia notoria entre el estado de reposo y peque√±os movimientos en el brazo con un mayor est√≠mulo como una contracci√≥n muscular.
